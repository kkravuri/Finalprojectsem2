```python
import os
import shutil

def copy_images_and_captions(source_folder, dest_folder, num_images_to_copy):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # List all files in the source Images folder
    image_files = os.listdir(os.path.join(source_folder, 'Images'))

    # Sort the image files based on their names or any specific criteria
    sorted_image_files = sorted(image_files)  # You might need a custom sorting logic here

    # Read the captions from the original captions.txt file
    with open(os.path.join(source_folder, 'captions.txt'), 'r') as captions_file:
        captions_lines = captions_file.readlines()

    # Open a new captions file for writing
    with open(os.path.join(dest_folder, 'new_captions.txt'), 'w') as new_captions_file:
        # Iterate over the sorted image files
        for image_filename in sorted_image_files[:num_images_to_copy]:
            # Find all matching captions for the current image
            matching_captions = [line.strip().split(',', 1)[1] for line in captions_lines if image_filename in line]
            print('matching_caption_line',matching_captions)

            if matching_captions:
                # Construct the source and destination paths for each image
                source_image_path = os.path.join(source_folder, 'Images', image_filename)
                dest_image_path = os.path.join(dest_folder, image_filename)

                # Copy the image to the new folder
                shutil.copyfile(source_image_path, dest_image_path)

                # Write all matching captions to the new captions file
                for caption in matching_captions:
                    new_captions_file.write(f"{image_filename},{caption}\n")
            else:
                print(f"Warning: Captions not found for image '{image_filename}'!")

if __name__ == "__main__":
    source_folder = 'flickr'
    dest_folder =os.path.join('CaptionDataSet_1000', 'images') 
    num_images_to_copy = 1000

    copy_images_and_captions(source_folder, dest_folder, num_images_to_copy)

    print(f"{num_images_to_copy} images and captions copied successfully.")
```
